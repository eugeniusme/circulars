```python
# процедура Алгоритма Дейкста, возвращает список путей
def Deikstra_route(node_count, gen_count):
    matr_smej  = [[0 for j in range(node_count)] for i in range(node_count)]

    d = []
    v = []

    for i in range(0, len(gen_count)):
        for j in range(0,node_count):
            k = gen_count[i] + j
            if k >= node_count:
                k = k - node_count
            #print('i=%d, j=%d, k=%d' % (i, j, k))
            matr_smej[j][k] = 1
            matr_smej[k][j] = 1
            #print(matr_smej)
            
    
    print(matr_smej)
    
    for i in range(node_count):
        d.append(10000)
        v.append(1)
    
    d[0] = 0
    
    # Шаг алгоритма
    minindex = 0
    while minindex < 10000:
        minindex = 10000
        min_ = 10000
        for i in range(node_count):
            if v[i] == 1 and d[i] < min_: #Если вершину ещё не обошли и вес меньше min, то значение обновляется
                
                min_ = d[i]
                minindex = i
            
            # Добавляем найденный минимальный вес к текущему весу вершины и сравниваем с текущим минимальным весом вершины
            if minindex != 10000:
                for i in range(node_count):
                    if matr_smej[minindex][i] > 0:
                        temp = min_ + matr_smej[minindex][i]
                        if temp < d[i]:
                            d[i] = temp
                v[minindex] = 0 
    #print(v, d)
    #конец алгоритма Дейкстры
    
    #Востановление пути
    ways = []
    
    for i in range(node_count):
        ver = [0] * node_count
        end = node_count - i - 1
        ver[0] = end
        k = 0 
        weight = d[end]
        
        while end > 0: 
            for j in range(node_count): 
                if matr_smej[end][j] != 0: 
                    temp = weight - matr_smej[end][j]
                    if temp == d[j]: 
                        weight = temp 
                        end = j 
                        #print('k=',k)
                        ver[k] = j + 1
                        k += 1
        ways.append(ver)
        
    min_ways_daikstra = []
        
    print(ways)
        
    for i in range(len(ways) - 1,-1,-1):
        buff = []
        for j in range(node_count - 1,-1,-1):
            #print(i,j)
            if ways[i][j] != 0:
                buff.append(ways[i][j])
        buff.append(len(min_ways_daikstra) + 1)
        min_ways_daikstra.append(buff)
        

    diam_graph = 0
    for i in range(len(min_ways_daikstra)):
        if len(min_ways_daikstra[i]) > diam_graph:
            diam_graph = len(min_ways_daikstra[i])
        diam_graph = diam_graph - 1
       
        
    return min_ways_daikstra
```

```python
# процедура расчета Коэффициента Эффективности
def Efficiency_algorithm(deikstra_routes, algorithm):
    deikstra_L = 0
    routing_algorithm_L = 0
    efficiency = 0

    for i in range(len(deikstra_routes)):
        deikstra_L += len(deikstra_routes[i]) - 1
        routing_algorithm_L += len(algorithm[i]) - 1

        
    efficiency = deikstra_L / routing_algorithm_L
    return efficiency
```

```python
# Алгоритм 1
def alg1(start_node, end_node, n, s1, s2, s3):
    if start_node > end_node:
        start_node = start_node - Step(end_node, start_node, n, s1, s2, s3)
    else:
        start_node = start_node + Step(start_node, end_node, n, s1, s2, s3)
    if start_node > n:
        start_node = start_node - n
    else:
        if start_node <= 0:
            start_node = start_node + n

    return start_node


def Step(start_node,  end_node,  N,  s1 = 1,  s2 = 2, s3 = 0):
    best_way_R = 100000, best_way_L = 100000
    step_R = 0, step_L = 0
    
    s = end_node - start_node

    R1 = (s) // s2 + s % N
    R2 = (s) // s2 - ( s % s2 ) + s2 + 1
    R3 = (s) // s2 - abs(( s2 * (s // s2 + 1) - s)) + s2 + 1
    R4 = (s) // s3 - ( s % s3 ) + s3 + 1
    R5 = (s) // s3 - abs(( s3 * (s // s3 + 1) - s )) + s3 + 1
    if (s % s3 == 0):
        return s3
    else:
        mass = [R1, R2, R3, R4, R5]
        r = mass.index(min(mass))
        
        if r == 0:
            best_way_R = R1
            step_R = s1
        elif r == 1:
            best_way_R = R2
            step_R = s2
        elif r == 2:
            best_way_R = R3
            step_R = s2
        elif r == 3:
            best_way_R = R4
            step_R = s3
        else: 
            best_way_R = R5
            step_R = s3

    s = start_node - end_node + N
    
    L1 = (s) // s2 + s % N
    L2 = (s) // s2 - ( s % s2 ) + s2 + 1
    L3 = (s) // s2 - abs((s2 * (s // s2 + 1) - s)) + s2 + 1
    L4 = (s) // s3 - ( s % s3 ) + s3 + 1
    L5 = (s) // s3 - abs((s3 * (s // s3 + 1) - s)) + s3 + 1
    
    if (s % (-s3) == 0):
        return -s3
    else:
        mass = [L1, L2, L3, L4, L5]
        
        r = mass.index(min(mass))

        if r == 0:
            best_way_L = L1
            step_L = -s1
        elif r == 1:
            best_way_L = L2
            step_L = -s2
        elif r == 2:
            best_way_L = L3
            step_L = -s2
        elif r == 3:
            best_way_L = L4
            step_L = -s3
        else: 
            best_way_L = L5
            step_L = -s3
    if (best_way_R < best_way_L):
        return step_R
    else:
        return step_L
```
